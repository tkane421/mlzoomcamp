```python
X = (df[df['Make'] == 'Rolls-Royce'])[['Engine HP','Engine Cylinders','highway MPG']].drop_duplicates().values
X
```

```python
XT = X.T
XT
XTX = XT.dot(X)
XTX
```

```python
XTXinv = np.linalg.inv(XTX)
XTXinv
```

```python
XTXinv.sum()
```
